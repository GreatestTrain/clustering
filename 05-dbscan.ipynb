```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs, make_circles
from sklearn.metrics import silhouette_score
```

```python
X, y = make_circles((100,200), noise=10)
```

```python
def plot_blobs(X, y, c=None, ax=None):
    if ax is None:
        ax = plt.gca()
    plots = []
    plots.append(sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, legend=True, style=y, ax=ax))
    if c not None:
        plots.append(ax.scatter(x=c[:,0], y=c[:,1], color='yellow', edgecolors='black', s=100, marker="X"))
    return plots
```

```python
plot_blobs(X,y)
```
