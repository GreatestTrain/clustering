```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
def plot_blobs(X, y, c=None, ax=None):
    if ax is None:
        ax = plt.gca()
    plots = []
    plots.append(sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, legend=True, style=y, ax=ax))
    if c:
        plots.append(ax.scatter(x=c[:,0], y=c[:,1], color='yellow', edgecolors='black', s=100, marker="X"))
    return plots
```

```python
X, y, c = make_blobs(500, cluster_std=0.8, centers=4, n_features=2, return_centers=True, random_state=42)
```

```python
dendogram_plot = dendrogram(linkage(X, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
model.fit(X)
y_pred = model.labels_
plot_blobs(X,y_pred)
plt.show()
```

```python
silhouette_score(X, y_pred)
```
