```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from matplotlib.colors import Colormap
```

```python
x, y = make_blobs(n_samples=100, centers=4, n_features=2, cluster_std=[1,1,2,2], random_state=7)
```

```python
import matplotlib as mpl
mpl.cm.get_cmap('viridis')(0)
```

```python
plt.figure(figsize=(6,6))
def plot_blobs(x, y, ax, cmap='viridis'):
    labels = np.unique(y)
    cmap_ = plt.get_cmap(cmap, lut=len(labels))
    for label in labels:
        sub_idx = np.argwhere(y == label).ravel()
        sub_x = x[sub_idx]
        sub_y = y[sub_idx]
        ax.scatter(sub_x[:,0], sub_x[:,1], color=cmap_(label), label=label)

plot_blobs(x,y, plt.gca(), cmap='Dark2')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

```python
kmeans = KMeans(n_clusters=4, random_state=7, n_init=1)
kmeans.fit(x)
pred = kmeans.predict(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(8,4))
plot_blobs(x,y, ax=axes[0], cmap='Dark2')
plot_blobs(x,pred, ax=axes[1], cmap='Dark2')
```
